```python
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import whisper
import torch
import librosa
import csv
import os




def transcribe_whisper(audio_path):
    # Load the model
    model = whisper.load_model("base")  # You can choose other models like "small", "medium", or "large"

    # Load audio and run the model
    result = model.transcribe(audio_path)
    
    # Return the transcription
    return result['text']

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")

def transcribe_wav2vec(audio_path):
    speech, rate = librosa.load(audio_path, sr=16000)
    input_values = processor(speech, return_tensors="pt", sampling_rate=16000).input_values
    with torch.no_grad():
        logits = model(input_values).logits
    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = processor.batch_decode(predicted_ids)[0]
    return transcription


original = "data_FinalProj\\original"
noise = "data_FinalProj\\noise"
csv_file_path = "masterfile.csv"

with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Add 'Wav2Vec' to the header row
    writer.writerow(['Filename', 'Whisper','Wav2Vec','Noise_Whisper','Noise_Wav2Vec'])  

    # Loop through each file in the directory
    for filename in os.listdir(original):
        if filename.endswith(".wav"):
            # Full path to the audio file
            original_path = os.path.join(original, filename)
            noise_path = os.path.join(noise, filename)
            
            whisper_transcription = transcribe_whisper(original_path)
            wav2vec_transcription = transcribe_wav2vec(original_path)
            whisper_noise_transcription = transcribe_whisper(noise_path)
            wav2vec_noise_transcription = transcribe_wav2vec(noise_path)

            # Write the filename and both transcriptions to CSV
            writer.writerow([filename, whisper_transcription])
            print(f"Processed {filename}")

print("All transcriptions have been saved to", csv_file_path)
```
